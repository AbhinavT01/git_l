```python
# # Import necessary libraries
# import pandas as pd
# import numpy as np
# import seaborn as sns
# import matplotlib.pyplot as plt
# from datetime import datetime
```

```python
# Load Datasets
# --------------------------------------------------------------------------------
customers = pd.read_csv("../Customers.csv")
products = pd.read_csv("../Products.csv")
transactions = pd.read_csv("../Transactions.csv")
```

```python
# Convert date columns to datetime format
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])
```

```python
# Merge datasets for relational insights
merged_data = pd.merge(transactions, customers, on="CustomerID")
merged_data = pd.merge(merged_data, products, on="ProductID")
```

```python
merged_data.rename(columns={'Price_x': 'ProductPrice', 'Price_y': 'TransactionPrice'}, inplace=True)
```

```python
# Check for missing values
print("Missing Values:\n")
print(merged_data.isnull().sum())
```

```python
# Basic statistics
print("\nBasic Statistics:\n")
print(merged_data.describe())
```

```python
# Unique customers, products, and transactions
print("\nUnique Customers:", customers['CustomerID'].nunique())
print("Unique Products:", products['ProductID'].nunique())
print("Unique Transactions:", transactions['TransactionID'].nunique())
```

```python
# Top 5 Products by Quantity Sold
top_products = merged_data.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 Products by Quantity Sold:\n", top_products)
```

```python
# Revenue Contribution by Region
region_revenue = merged_data.groupby('Region')['TotalValue'].sum().sort_values(ascending=False)
print("\nRevenue by Region:\n", region_revenue)
```

```python
# Revenue by Category and Region
category_region_revenue = merged_data.groupby(['Region', 'Category'])['TotalValue'].sum().unstack()
print("\nRevenue by Category and Region:\n", category_region_revenue)
```

```python
# Popular Products by Region
popular_products_region = merged_data.groupby(['Region', 'ProductName'])['Quantity'].sum().reset_index()
popular_products_region = popular_products_region.sort_values(['Region', 'Quantity'], ascending=[True, False])
print("\nTop Products by Region:\n", popular_products_region.groupby('Region').head(3))
```

```python
# Customer Signup and Purchase Trends
merged_data['SignupToPurchaseLag'] = (merged_data['TransactionDate'] - merged_data['SignupDate']).dt.days
print("\nAverage Signup to Purchase Lag (Days):", merged_data['SignupToPurchaseLag'].mean())
```

```python
# Revenue Contribution by Customer
customer_revenue = merged_data.groupby('CustomerID')['TotalValue'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Customers by Revenue Contribution:\n", customer_revenue)
```

```python
#montly revenue trends
monthly_revenue = merged_data.groupby(merged_data['TransactionDate'].dt.to_period('M'))['TotalValue'].sum()


monthly_revenue_df = monthly_revenue.reset_index()


print(monthly_revenue_df)
```

```python
# Revenue Trends Over Time
monthly_revenue = merged_data.groupby(merged_data['TransactionDate'].dt.to_period('M'))['TotalValue'].sum()
monthly_revenue.plot(kind='line', figsize=(12, 6), marker='o', title="Monthly Revenue Trends")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.show()
```

```python
# Grouping the data by Region and Category, then summing the total revenue
region_category_revenue = merged_data.groupby(['Region', 'Category'])['TotalValue'].sum().unstack()

# Plotting the revenue by Region and Category as a bar plot
region_category_revenue.plot(kind='bar', figsize=(10, 6), stacked=False, title="Revenue by Region and Category")
plt.xlabel("Region")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.legend(title="Category")
plt.show()
```

```python
# Grouping the data by Region and Product, then summing the quantity sold
region_product_quantity = merged_data.groupby(['Region', 'ProductName'])['Quantity'].sum().unstack()

# Sorting the products by quantity for each region and selecting top 5 products for each region
region_product_quantity = region_product_quantity.apply(lambda x: x.sort_values(ascending=False).head(5), axis=1)

# Grouping the data for overall top products by summing quantities across all regions
overall_product_quantity = merged_data.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False).head(5)

# Plotting the top products for each region
ax = region_product_quantity.plot(kind='bar', figsize=(14, 6), title="Top Products by Quantity Sold for Each Region and Overall")
ax.set_xlabel("Region")
ax.set_ylabel("Quantity Sold")
ax.set_xticklabels([f"{region}\n{product}" for region in region_product_quantity.index for product in region_product_quantity.columns], rotation=45)

# Plotting the overall top products as a separate set of bars in the same figure
overall_product_quantity.plot(kind='bar', ax=ax, color='orange', position=1, width=0.4, legend=True)

plt.legend(title="Product Name", loc='upper left')
plt.show()
```

```python
# 5. Advanced Insights
# --------------------------------------------------------------------------------
# Cohort Analysis: Signup Trends
customers['SignupMonthYear'] = customers['SignupDate'].dt.to_period('M')
cohort = merged_data.groupby(customers['SignupMonthYear'])['CustomerID'].nunique()
print("\nCohort Analysis (Unique Customers per Signup Month):\n", cohort)
```

```python

# Monthly Popular Products
monthly_popular = merged_data.groupby([merged_data['TransactionDate'].dt.to_period('M'), 'ProductName'])['Quantity'].sum().reset_index()
monthly_top_products = monthly_popular.sort_values(['TransactionDate', 'Quantity'], ascending=[True, False]).groupby('TransactionDate').head(3)
print("\nMonthly Top Products:\n", monthly_top_products)
```

```python
# Grouping the data by month and category, then summing the total revenue
category_trends = merged_data.groupby([merged_data['TransactionDate'].dt.to_period('M'), 'Category'])['TotalValue'].sum().unstack()

# Convert the result into a DataFrame for better readability
category_trends_df = category_trends.reset_index()

# Display the tabular form of category sales trends
print(category_trends_df)
```

```python
# Category Sales Trends Over Time
category_trends = merged_data.groupby([merged_data['TransactionDate'].dt.to_period('M'), 'Category'])['TotalValue'].sum().unstack()
category_trends.plot(kind='line', figsize=(12, 6), marker='o', title="Category Sales Trends Over Time")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.legend(title="Category")
plt.show()

```

```python
# High-Spending Customers
high_spending_customers = merged_data.groupby('CustomerID')['TotalValue'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 High-Spending Customers:\n", high_spending_customers)
```

```python
# Products with Highest Margins
product_margin = merged_data.groupby('ProductName').apply(lambda x: (x['ProductPrice'].mean() - x['TotalValue'].sum()/x['Quantity'].sum()) / x['ProductPrice'].mean() * 100).sort_values(ascending=False).head(5)
print("\nTop 5 High-Margin Products:\n", product_margin)
```

```python
# Region Contribution to High-Spending Customers
region_high_spending = merged_data[merged_data['CustomerID'].isin(high_spending_customers.index)].groupby('Region')['TotalValue'].sum()
print("\nRegion Contribution to High-Spending Customers:\n", region_high_spending)
```

```python
#Customer Retention Rate
retention_rate = merged_data.groupby('CustomerID')['TransactionDate'].nunique().reset_index()
retention_rate.columns = ['CustomerID', 'PurchaseFrequency']
average_retention_rate = retention_rate['PurchaseFrequency'].mean()
print("\nAverage Customer Retention Rate (Frequency):", average_retention_rate)
```

```python
# Grouping by Region and CustomerID, then calculating the number of unique transactions for each customer
region_retention_rate = merged_data.groupby(['Region', 'CustomerID'])['TransactionDate'].nunique().reset_index()

# Renaming the columns for clarity
region_retention_rate.columns = ['Region', 'CustomerID', 'PurchaseFrequency']

# Calculating the average purchase frequency (retention rate) for each region
region_average_retention_rate = region_retention_rate.groupby('Region')['PurchaseFrequency'].mean()

# Printing the retention rate for each region
print("\nAverage Customer Retention Rate by Region:")
print(region_average_retention_rate)

```

```python

# Product Repeatability
product_repeatability = merged_data.groupby('ProductName')['CustomerID'].nunique() / merged_data['CustomerID'].nunique()
product_repeatability = product_repeatability.sort_values(ascending=False).head(5)
print("\nTop 5 Most Repeatedly Bought Products:\n", product_repeatability)
```

```python
# Grouping the data by Region, ProductName, and Category, then summing the quantity sold
product_region_popularity = merged_data.groupby(['Region', 'ProductName', 'Category'])['Quantity'].sum().reset_index()

# Sorting by Region and Quantity (in descending order), and then selecting the top 3 products per region
product_region_popularity = product_region_popularity.sort_values(['Region', 'Quantity'], ascending=[True, False]).groupby('Region').head(3)

# Display the top products along with their categories in each region
print("\nTop Products in Each Region with Category:\n", product_region_popularity)
```

```python
# Grouping the data by Region and Category, then summing the total quantity sold for each category
category_sales_by_region = merged_data.groupby(['Region', 'Category'])['Quantity'].sum().reset_index()

# Sorting the data by Region and Sales (in descending order) to get the top categories
category_sales_by_region = category_sales_by_region.sort_values(['Region', 'Quantity'], ascending=[True, False])

# Selecting the top 2 selling categories in each region
top_2_categories_per_region = category_sales_by_region.groupby('Region').head(2)

# Displaying the top 2 categories in each region
print("\nTop 2 Categories in Each Region:\n", top_2_categories_per_region)
```

```python
# Revenue by Product Category
category_revenue = merged_data.groupby('Category')['TotalValue'].sum().sort_values(ascending=False)
category_revenue.plot(kind='bar', figsize=(10, 6), title="Revenue by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.show()
```

```python
# Grouping the data by Region and Category, and then summing the total revenue (TotalValue) for each category in each region
category_revenue_by_region = merged_data.groupby(['Region', 'Category'])['TotalValue'].sum().reset_index()

# Sorting the results by Region and Revenue (in descending order)
category_revenue_by_region = category_revenue_by_region.sort_values(['Region', 'TotalValue'], ascending=[True, False])

# Plotting the revenue by category in each region
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.barplot(data=category_revenue_by_region, x='Category', y='TotalValue', hue='Region')
plt.title("Revenue by Product Category in Each Region")
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Grouping the data by Region and Category, and then summing the total revenue (TotalValue) for each category in each region
category_revenue_by_region = merged_data.groupby(['Region', 'Category'])['TotalValue'].sum().reset_index()

# Sorting the results by Region and Revenue (in descending order)
category_revenue_by_region = category_revenue_by_region.sort_values(['Region', 'TotalValue'], ascending=[True, False])

# Display the results in tabular form
print("\nRevenue by Product Category in Each Region:\n", category_revenue_by_region)

# Plotting the revenue by category in each region
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.barplot(data=category_revenue_by_region, x='Category', y='TotalValue', hue='Region')
plt.title("Revenue by Product Category in Each Region")
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Customer Purchase Frequency Distribution
retention_rate['PurchaseFrequency'].plot(kind='hist', bins=20, figsize=(10, 6), title="Customer Purchase Frequency Distribution")
plt.xlabel("Number of Purchases")
plt.ylabel("Frequency")
plt.show()
```

```python
# Revenue Contribution by High and Low Spending Customers
threshold = merged_data['TotalValue'].median()
high_spending = merged_data[merged_data['TotalValue'] > threshold].groupby('CustomerID')['TotalValue'].sum().sum()
low_spending = merged_data[merged_data['TotalValue'] <= threshold].groupby('CustomerID')['TotalValue'].sum().sum()
spending_data = pd.Series([high_spending, low_spending], index=['High Spending', 'Low Spending'])
spending_data.plot(kind='bar', figsize=(8, 5), title="Revenue Contribution: High vs Low Spending Customers")
plt.ylabel("Total Revenue")
plt.show()
```

```python
# Top Products by Profit Margin
top_margin_products = merged_data.groupby('ProductName').apply(
    lambda x: (x['ProductPrice'].mean() - x['TotalValue'].sum()/x['Quantity'].sum()) / x['ProductPrice'].mean() * 100
).sort_values(ascending=False).head(5)

top_margin_products.plot(kind='bar', figsize=(10, 6), title="Top 5 Products by Profit Margin")
plt.xlabel("Product Name")
plt.ylabel("Profit Margin (%)")
plt.show()
```

```python
category_region_revenue = merged_data.groupby(['Region', 'Category'])['TotalValue'].sum().unstack()
plt.figure(figsize=(12, 8))
sns.heatmap(category_region_revenue, annot=True, fmt=".2f", cmap="coolwarm", linewidths=.5)
plt.title("Revenue by Category and Region")
plt.show()
```

```python
# Calculate total profit for each product
merged_data['Profit'] = (merged_data['ProductPrice'] * merged_data['Quantity']) - merged_data['TotalValue']
product_profit_sales = merged_data.groupby('ProductName').agg({'Profit': 'sum', 'Quantity': 'sum'}).sort_values(by='Profit', ascending=False).head(5)
print("\nTop 5 Products by Profit and Sales:\n", product_profit_sales)
```

```python
# Visualize Top 5 Products by Sales
product_profit_sales['Quantity'].plot(kind='bar', figsize=(10, 6), title="Top 5 Products by Sales")
plt.xlabel("Product Name")
plt.ylabel("Quantity Sold")
plt.show()
```
